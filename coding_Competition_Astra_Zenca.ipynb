```python
# st= (str)(df['openfda'][0])
# st.split(",")
import re
route=[]
drugname=[]
for i in range(len(l)):
    s= (str)(df['openfda'][i])
    k=s.split(',')
    for j in k:
        if( re.search('route.+', j)):
            match=re.search('route.+', j)
            route.append(match.group())
        if(re.search('substance_name.+', j)):
            match1=re.search('substance_name.+', j)
            drugname.append(match1.group())
```
